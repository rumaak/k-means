# K-means clustering

We will write our own implementation of K-means algorithm. As a training set, we will use synthetic sklearn dataset.

```python
%matplotlib inline

from matplotlib import pyplot as plt 
from sklearn.datasets import make_blobs
import sklearn.datasets
import numpy as np
```

### Data

```python
xs,ys = make_blobs()
```

```python
plt.scatter(xs[:,0],xs[:,1], c=ys)
```

### Model

```python
class K_means_clf(object):
    
    def __init__(self):
        self.K = None
        self.cents = None
        
    def train(self, data, K, epsilon=1e-5):
        '''
        Iterates over data and updates centroids until convergence
        ''' 
        self.K = K
        costs = [self.cost(data)+epsilon, self.cost(data)]
        print(costs[-1])
        while((costs[-2]-costs[-1])>=epsilon):
            x_to_class = self.forward(data)
            split_to_class = [np.where(x_to_class==c) for c in range(0,self.K)]
            self.cents = [np.mean(data[idxs],axis=0) for idxs in split_to_class]
            costs.append(self.cost(data))
            print(costs[-1])
    
    def forward(self,data):
        '''
        Forward pass of K-means algorithm
        '''
        new_data, new_cent = self.initialize(data) 
        dists = self.dist(new_data, new_cent)
        x_to_class = np.argmin(dists, axis=0)
        return x_to_class
    
    def dist(self, v1, v2):
        '''
        Euclidian distance of two vectors (assuming coordinates are last dimension of input arrays)
        '''
        return np.sqrt(np.sum(((v1-v2)**2),axis=-1))
    
    def __add_dims(self, xs, cents):
        '''
        Adds dimension for numpy broadcasting.
        '''
        new_data = np.reshape(xs,(1,*np.shape(xs)))
        new_cent = np.reshape(cents,(np.shape(cents)[0], 1, np.shape(cents)[1]))
        return new_data, new_cent
    
    def distortion(self, xs, cents):
        '''
        Optimization objective. Defines how close are examples to their respective centroids.
        '''
        return (np.sum(self.dist(xs,cents)**2,axis=0))/np.shape(xs)[0]
    
    def initialize(self, data):
        '''
        Creates (if not already created) centroids and adds dimensions for broadcasting
        '''
        if self.cents is not None:
            new_data, new_cent = self.__add_dims(data,self.cents)
            return new_data, new_cent
        idxs = np.random.randint(low=0, high=len(data), size=self.K)
        self.cents = data[idxs]
        return self.__add_dims(data,self.cents)
        
    def get_corr_cents(self,x_to_class):
        '''
        Returns an array with corresponding centroid vectors
        '''
        corresponding_cents = np.zeros((len(x_to_class),np.shape(self.cents)[1]))
        for i,c in enumerate(self.cents):
            mask = (x_to_class==i)[:,None]
            np.copyto(corresponding_cents,c,where=mask) 
        return corresponding_cents
    
    def cost(self,data):
        '''
        Prepares data and uses loss function on it
        '''
        x_to_class = self.forward(data)
        corr_cents = self.get_corr_cents(x_to_class)
        return self.distortion(data,corr_cents)
```

```python
K = 3
keke = K_means_clf()
keke.train(xs,K)
```

```python
tmp_xs = np.concatenate((xs,keke.cents))
tmp_ys = np.concatenate((ys,[3]*K))
```

```python
plt.scatter(tmp_xs[:,0],tmp_xs[:,1], c=tmp_ys)
```
